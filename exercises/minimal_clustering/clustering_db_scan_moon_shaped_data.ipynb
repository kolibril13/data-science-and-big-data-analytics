```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
plt.rcParams['figure.dpi'] = 150  
```

```python
from sklearn.datasets import make_moons
Xmoon, ymoon = make_moons(200, noise=.05, random_state=0)
plt.scatter(Xmoon[:, 0], Xmoon[:, 1]);
```

```python
from sklearn.cluster import KMeans
wss= []
index = []
for i in range(1,10):
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=10000, 
        tol=1e-04, random_state=0
    )
    y_km = km.fit_predict(Xmoon)
    wss.append(km.inertia_)
    index.append(i)
index, wss
plt.plot(index,wss, marker= "+" )
```

```python
km = KMeans(
        n_clusters=2, init='random',
        n_init=10, max_iter=10000, 
        tol=1e-04, random_state=0
    )
y_km = km.fit_predict(Xmoon)
plt.scatter(Xmoon[:,0], Xmoon[:,1], c=y_km, s=50, cmap=plt.cm.Paired, alpha=0.4)
plt.scatter(km.cluster_centers_[:, 0],km.cluster_centers_[:, 1], 
        s=250, marker='*', label='centroids',
        edgecolor='black',
       c=np.arange(0,2),cmap=plt.cm.Paired,)
```

```python
Xmoon=np.append(Xmoon, [[-1,-1]], axis=0)
```

```python
from sklearn.cluster import DBSCAN
from ipywidgets import *
def func(f):
    db = DBSCAN(eps=f, min_samples=7).fit(Xmoon)
    dblabels = db.labels_
    plt.scatter(Xmoon[:,0], Xmoon[:,1], c=dblabels, s=50, cmap=plt.cm.Paired, alpha=0.4)
interact(func,f=FloatSlider(min=0.1, max=2, step=0.1,continuous_update=False)); 
```
