```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
plt.rcParams['figure.dpi'] = 150  
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# create dataset
X, y = make_blobs(
   n_samples=400, n_features=2,
   centers=1, cluster_std=1, center_box= (0,0),
   shuffle=False, random_state=0
)
X=X.T
# plot
plt.scatter( X[0], X[1], s=50, alpha=0.4)
plt.axis([-10,10,-10,10])
```

Normale Verteilung ergibt folgende Verteilung
$$ \Sigma = 
  \begin{bmatrix}
    1 & 0\\
    0 & 1
    \end{bmatrix}
 $$

```python
np.set_printoptions(suppress=True)
np.cov(X)
```

```python
T = np.array([[4,0],[0,1]])
X2= T @ X
```

```python
plt.scatter( X2[0], X2[1], s=50, alpha=0.4)
plt.axis([-10,10,-10,10])
```

Nach der Transformation:
$$ \Sigma = 
  \begin{bmatrix}
    16 & 0\\
    0 & 1
    \end{bmatrix}
 $$

```python
np.cov(X2)
```

```python
theta=np.pi/2*1.1
r = np.array(( (np.cos(theta), -np.sin(theta)),
               (np.sin(theta),  np.cos(theta)) ))
 
X3= r @ X2
plt.scatter( X3[0], X3[1], s=50, alpha=0.4)
plt.axis([-10,10,-10,10])
```

```python
matrix=np.cov(X3)
matrix
```

```python
w, v = np.linalg.eig(matrix)
w= np.sqrt(w)
w
```

```python
v
```

```python
plt.arrow(x=0,y=0,dx=v[0,0]*w[0],dy=v[1,0]*w[0] , lw=3, color="r")
plt.arrow(x=0,y=0,dx=v[0,1]*w[1],dy=v[1,1]*w[1] , lw=3, color="r")

plt.scatter( X3[0], X3[1], s=50, alpha=0.4)
plt.axis([-10,10,-10,10])
```
