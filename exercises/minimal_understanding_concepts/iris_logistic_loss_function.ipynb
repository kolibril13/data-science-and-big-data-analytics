```python
def sigmoid(x):
    # Activation function used to map any real value between 0 and 1
    return 1 / (1 + np.exp(-x))

def net_input(theta, x):
    # Computes the weighted sum of inputs
    return np.dot(x, theta)


def probability(theta, x):
    # Returns the probability after passing through sigmoid
    return sigmoid(net_input(theta, x))

def cost_function(theta, x, y):
    # Computes the cost function for all the training samples
    m = x.shape[0]
    total_cost = -(1 / m) * np.sum(
        y * np.log(probability(theta, x)) + (1 - y) * np.log(
            1 - probability(theta, x)))
    return total_cost

from ipywidgets import *
def func(t1,t2):
    teta0= t1
    teta1= t2
    theta = np.array([teta0,teta1])
    theta=np.reshape(theta, (2,-1))
    x_ones=np.ones(( len(X),1))
    X.shape, x_ext.shape
    X2= np.append(x_ones,X, axis=1)


    #print(probability(theta, X2))
    cost= cost_function(theta, X2, y.T)
    print(cost)
    X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
    y_own= sigmoid(teta0+teta1*X_new)
    plt.plot(X_new, y_own)
    plt.plot(X_new, y_proba[:, 1],  "r", linewidth=40, alpha=0.1)
interact(func,t1=FloatSlider(1.1,min=-10, max=10, step=0.1,continuous_update=False), t2=FloatSlider(1.1,min=0, max=9, step=0.1,continuous_update=False)); 
print("best= t1=-3.2, t2=1.9")
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(solver="liblinear", random_state=42)
log_reg.fit(X, y)
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)

plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Not Iris-Virginica")
plt.plot(X_new, 1-y_proba[:, 0], "r", linewidth=10, alpha=0.1, label="Not Iris-Virginica")
```
