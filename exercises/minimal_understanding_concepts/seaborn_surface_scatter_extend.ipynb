```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set()
from sklearn import datasets
```

```python
iris = datasets.load_iris()
plt.rcParams['figure.dpi'] = 150  
X= pd.DataFrame(iris.data, columns= iris.feature_names)
X = X.iloc[:, [0,1]]
y = iris.target
iris.feature_names
```

```python
my_dict = {0:'Iris-setosa', 1:'Iris-versicolor', 2:'Iris-virginica'}
y_l = [my_dict[yi] for yi in y]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
clf3 = RandomForestClassifier(max_depth=2, random_state=0)
clf3.fit(X, y)

print(clf3.feature_importances_)

print(clf3.predict([[0, 0]]))
```

```python
grid_dist=0.05
x_min= X.iloc[:, 0].min() - 1 ; x_max= X.iloc[:, 0].max() + 1
y_min= X.iloc[:, 1].min() - 1 ; y_max= X.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, grid_dist),
                     np.arange(y_min, y_max, grid_dist))
Z=clf3.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
plt.subplots(1, 1, figsize=(10, 8))
plt.rcParams['figure.dpi'] = 100  
plt.contourf(xx, yy, Z, alpha=1, cmap=plt.cm.RdYlBu)
sns.scatterplot(X.iloc[:, 0], X.iloc[:, 1], hue= y_l , s=100, edgecolor='black')
```
