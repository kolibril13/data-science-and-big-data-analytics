```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set()
from sklearn import datasets
```

```python
iris = datasets.load_iris()
plt.rcParams['figure.dpi'] = 150  
X= pd.DataFrame(iris.data, columns= iris.feature_names)
X = X.iloc[:, [0,1]]
y = iris.target
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf2 = KNeighborsClassifier(n_neighbors=8)
clf2.fit(X, y)
```

```python
x_min= X.iloc[:, 0].min() - 1 ; x_max= X.iloc[:, 0].max() + 1
y_min= X.iloc[:, 1].min() - 1 ; y_max= X.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.04),
                     np.arange(y_min, y_max, 0.04))
Z=clf2.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
plt.subplots(1, 1, figsize=(10, 8))
plt.rcParams['figure.dpi'] = 100  
plt.contourf(xx, yy, Z, alpha=1, cmap=plt.cm.RdYlBu)
sns.scatterplot(X.iloc[:, 0], X.iloc[:, 1], hue= y, palette=plt.cm.RdYlBu  , s=100, edgecolor='black')
```
