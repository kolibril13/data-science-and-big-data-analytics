```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd
plt.rcParams['figure.dpi'] = 150  
plt.rcParams['figure.dpi'] = 130 
```

```python
from sklearn.datasets import make_blobs
plt.rcParams['figure.dpi'] = 150 
# create dataset
X, y = make_blobs(
   n_samples=300, n_features=2,
   centers=12, cluster_std=0.2,
   shuffle=False, random_state=1
)


# plot
plt.scatter(
   X[:, 0], X[:, 1],
   edgecolor='black', s=50, c= y
)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from ipywidgets import *
def func(val):
    clf2 = KNeighborsClassifier(n_neighbors=int(val))
    clf2.fit(X, y)

    x_min= X[:, 0].min() - 1
    x_max= X[:, 0].max() + 1
    y_min= X[:, 1].min() - 1
    y_max= X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))
    Z=clf2.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    f, axarr = plt.subplots(1, 1, figsize=(10, 8))
    axarr.scatter(X[:, 0], X[:, 1], c=y,  s=100, edgecolor='k', cmap="Dark2")
    axarr.contourf(xx, yy, Z, alpha=0.4, cmap="Dark2")
interact(func,val=FloatSlider(min=1, max=23, step=1,continuous_update=False)); 
```
