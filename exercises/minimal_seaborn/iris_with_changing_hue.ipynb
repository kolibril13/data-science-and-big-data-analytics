```python
import pandas as pd
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import copy
plt.figure(figsize = (16,9))
plt.rcParams['figure.dpi'] = 150
data = sns.load_dataset("iris")
```

```python
data2= copy.deepcopy(data)
data2["pw_check"] = data2["petal_width"] > 0.2
data2["pw_check"]=data2["pw_check"].replace({False: "small", True: "big"})
```

```python
sns.swarmplot(data2["species"], data2["sepal_length"], hue=data2["pw_check"])
```
